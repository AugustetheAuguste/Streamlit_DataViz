```python
df = pd.read_json("Datasets\subset_78_all_years.jsonl", lines = True)
df.head()
```

```python
result = df.loc[(df["Nature mutation"] == "Vente") & (df["Valeur fonciere"] > 40000000)]
result
```

```python
# Définir la colonne "Valeur fonciere" comme la colonne à analyser
colonne_valeur_fonciere = df["Valeur fonciere"]

# Calculer le premier quartile (Q1) et le troisième quartile (Q3)
Q1 = colonne_valeur_fonciere.quantile(0.25)
Q3 = colonne_valeur_fonciere.quantile(0.75)

# Calculer l'écart interquartile (IQR)
IQR = Q3 - Q1

# Calculer les limites supérieure et inférieure pour les outliers
limite_superieure = Q3 + 1.5 * IQR
limite_inferieure = Q1 - 1.5 * IQR

# Filtrer les outliers
outliers = df[(colonne_valeur_fonciere < limite_inferieure) | (colonne_valeur_fonciere > limite_superieure)]
outliers.shape
```
